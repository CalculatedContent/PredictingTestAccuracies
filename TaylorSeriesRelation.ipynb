```python
import numpy as np
from tqdm import tqdm_notebook as tqdm

from sklearn.decomposition import TruncatedSVD
import powerlaw

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
mus = [0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3.0, 3.25, 3.5]
y = []
ts = []
for mu in mus:
    exp =(4/mu)-1
    y.append(1/exp)
    ts.append(mu/2)
```

```python
plt.scatter(mus,y, label=r"$\dfrac{1}{(4/\mu)-1}$")
plt.scatter(mus,ts, label=r"$\dfrac{\mu}{2}$")


plt.rcParams['figure.figsize'] = [8,8]
matplotlib.rcParams.update({'font.size': 20})
legend_prop={'size':18}
marker_size = 75

plt.xlabel(r"$\mu$")
plt.legend(prop=legend_prop)
plt.title(r"Taylor Series Relation for $\mu$")
plt.savefig("img/taylor-series.png")
```
